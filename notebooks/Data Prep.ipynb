```python
import pandas as pd
```

```python
# df = pd.DataFrame(columns=['time', 'y'])
df = pd.DataFrame()


for file in ['2020-01-03', '2020-04-06', '2020-07-09', '2020-10-12', '2021-01', '2021-02', '2021-03', '2021-04', '2021-05', '2021-06', '2021-07', '2021-08', '2021-09', '2021-10', '2021-11', '2021-12', '2022-01-01-09']:
    monthly_data = pd.read_csv(file+'.csv')
    columns = monthly_data.columns.to_list()
    columns.remove('dateflg')
    columns.remove('timeflg')
    data_list = []
    for i, row in monthly_data.iterrows():
        date = row['dateflg']
        time = row['timeflg']-1
#     total = row['Inflow_Total']
#     stretcher = row['INFLOW_STRETCHER']
#     ambulatory = row['INFLOW_AMBULATORY']
#     total_cum = row['Inflow_Cum_Total']
        time = pd.to_datetime(date + ' ' + str(time)+':00')
        dict_to_append = {}
        dict_to_append['time'] = time
#     data_list.append({'time':time, 'stretcher':stretcher, 'ambulatory':ambulatory,'total':total, 'total_cum':total_cum})
        for column in columns:
            dict_to_append[column] = row[column]
    
        data_list.append(dict_to_append)
        data_df = pd.DataFrame(data_list)
        df = pd.concat([df, data_df]).sort_values(by='time', ascending=True, ignore_index=True)
        df = df.drop_duplicates(keep='first')

df = df.set_index('time').resample('H').sum().reset_index()

df
```

```python
df.to_csv('concat.csv', index=False)
```

```python
df = pd.DataFrame()


for file in ['2021-01', '2021-02', '2021-03', '2021-04', '2021-05', '2021-06', '2021-07', '2021-08', '2021-09', '2021-10']:
    monthly_data = pd.read_csv(file+'.csv')
    df = pd.concat([df, monthly_data]).sort_values(by=['dateflg', 'timeflg'], ascending=True, ignore_index=True)
    df = df.drop_duplicates(keep='first')

df
```

```python
output = pd.DataFrame()
for date in df.dateflg.unique():
    target = df[df.dateflg==date][['Inflow_Cum_Total']].reset_index(drop=True)
    target.columns = [date]
    target = target.transpose()
    output = pd.concat([output, target])
output.index = pd.to_datetime(output.index)
output
```

```python
output.to_csv('hourly_total_cum_inflow.csv')
```
